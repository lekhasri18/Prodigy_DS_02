Load Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from the uploaded CSV file
file_path = "C:/Users/DELL/Downloads/2015.csv"
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(data.head())

# Display basic information about the dataset
print(data.info())
```

Data Cleaning

```python
# Check for missing values
print(data.isnull().sum())

# Drop rows with any missing values
cleaned_data = data.dropna()

# Verify that there are no more missing values
print(cleaned_data.isnull().sum())

# Display basic statistics
print(cleaned_data.describe())
```

Exploratory Data Analysis (EDA)


```python
# Set the style for the plots
sns.set(style="whitegrid")

# Plot the distribution of a continuous variable (e.g., Happiness Score)
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_data['Happiness Score'], bins=30, kde=True)
plt.title('Distribution of Happiness Score')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of another continuous variable (e.g., Economy (GDP per Capita))
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_data['Economy (GDP per Capita)'], bins=30, kde=True)
plt.title('Distribution of Economy (GDP per Capita)')
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the count of a categorical variable (e.g., Region)
plt.figure(figsize=(10, 6))
sns.countplot(y='Region', data=cleaned_data, palette='viridis', order=cleaned_data['Region'].value_counts().index)
plt.title('Count of Countries by Region')
plt.xlabel('Count')
plt.ylabel('Region')
plt.show()
```

```python
# Plot the relationship between Happiness Score and Economy (GDP per Capita)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Economy (GDP per Capita)', y='Happiness Score', data=cleaned_data, hue='Region', palette='viridis')
plt.title('Happiness Score vs Economy (GDP per Capita)')
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.show()
```

```python
# Plot the relationship between Happiness Score and Health (Life Expectancy)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Health (Life Expectancy)', y='Happiness Score', data=cleaned_data, hue='Region', palette='viridis')
plt.title('Happiness Score vs Health (Life Expectancy)')
plt.xlabel('Health (Life Expectancy)')
plt.ylabel('Happiness Score')
plt.show()
```

```python
# Plot a correlation heatmap for numeric columns
plt.figure(figsize=(12, 8))
numeric_columns = cleaned_data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = cleaned_data[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Plot pairplot of numerical variables
sns.pairplot(cleaned_data[['Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Freedom']])
plt.show()
```
